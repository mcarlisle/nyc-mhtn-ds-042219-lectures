```python
from scipy.stats import binom
p = 0.55
for n in range(50,101):
    majority_prob = sum([binom.pmf(k,n,p) for k in range(n//2 + 1,n + 1)])
    print(f"n = {n}: P(majority agree) = {majority_prob}")
    if majority_prob > 0.8:
        print(f"    {n} amateurs probabilistically score better than the expert!")
```

```python
p = 0.55
for n in range(50,101):
    majority_prob = 1 - binom.cdf(n // 2, n, p)
    print(f"n = {n}: P(majority agree) = {majority_prob}")
    if majority_prob > 0.8:
        print(f"    {n} amateurs probabilistically score better than the expert!")
```

```python
import random
from numpy import mean
import matplotlib.pyplot as plt

N = 10000
N_samples = list(range(N))
bootstrap_uniques = []

for m in range(10000):
    N_bootstrap = [random.choice(N_samples) for i in range(N)]
    set_N_bootstrap = set(N_bootstrap)
    bootstrap_uniques.append(len(set_N_bootstrap))
    
ave_bootstrap_size = mean(bootstrap_uniques)
print(ave_bootstrap_size)
plt.hist(bootstrap_uniques)
plt.show();
```

```python
def expected_coupon_collector_time(n):
    return round(1 + n * log(n), 3)

def expected_coupon_collector_time_sum(n):
    return round(1 + n * sum([1/i for i in range(1,n+1)]), 3)
```

```python
for i in range(1000,200001,1000):
    #print(f"{i}: {expected_coupon_collector_time(i * (1 - exp(-i)))}")
    pct_error = round((expected_coupon_collector_time_sum(i) - expected_coupon_collector_time(i))  \
                / expected_coupon_collector_time_sum(i), 5)
    print(f"{i}: {expected_coupon_collector_time_sum(i)}; {pct_error}% error")
```
